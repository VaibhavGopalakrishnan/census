The differences in census tract boundaries over time can be attributed to several factors related to population growth, urban development, and changes in administrative boundaries. Here's an explanation of how to interpret the changes in your data, based on the common names and unique values across the 2010 and 2020 census tract datasets:

### Key Findings:
1. **Common Census Tracts (334)**:
   - **Explanation**: These tracts remained unchanged between 2010 and 2020. The fact that you have 334 common tracts means that the majority of census tracts have maintained their boundaries and naming conventions over the decade.
   - **Interpretation**: This stability can be expected in areas where there hasn't been significant population growth or decline, and no major boundary redistricting has occurred. These areas likely have a stable population density or are in regions where geographic boundaries (like natural features) constrain changes.

2. **Unique Census Tracts in 2010 (38)**:
   - **Explanation**: These tracts existed in 2010 but no longer exist in the 2020 dataset. This could be due to:
     - **Merging of Tracts**: Some tracts may have been combined with others if population density decreased in these areas.
     - **Renaming**: The names of some tracts may have changed over time, though the geographic area may still exist under a new tract name.
     - **Reclassification**: These tracts may have been redefined, resulting in changes to the boundaries, which led to their removal or reassignment.
   - **Interpretation**: These changes may reflect urban decay, migration out of certain areas, or administrative decisions to streamline boundaries in rural or underpopulated areas.

3. **Unique Census Tracts in 2020 (74)**:
   - **Explanation**: These tracts are new in the 2020 dataset, meaning they did not exist in 2010. This could be due to:
     - **Newly Created Tracts**: As population increases in certain areas, particularly in urban or suburban developments, new census tracts may be created to ensure that each tract meets population thresholds set by the Census Bureau (typically between 1,200 to 8,000 people).
     - **Splitting of Existing Tracts**: Larger tracts in 2010 may have been split into smaller tracts to account for population growth or shifts in density.
     - **Administrative or Redistricting Changes**: Boundaries may be redefined due to local or federal administrative decisions, sometimes to reflect new political, economic, or environmental factors.
   - **Interpretation**: This likely indicates significant growth or changes in population distribution, particularly in areas experiencing urban sprawl, development, or population booms. These new tracts may also reflect finer granularity in data collection for areas that have become more densely populated.

### Factors That Influence Census Tract Boundary Changes:
- **Population Growth or Decline**: The Census Bureau aims to keep the population within each tract relatively consistent. Areas with rapid population growth (e.g., urban or suburban areas) may see large tracts split into smaller ones, while areas with population decline may see tracts merged.
  
- **Urban Development**: Expanding cities and towns can lead to the creation of new tracts, particularly if there is significant residential or commercial development. These areas are often divided into new tracts to ensure accurate demographic reporting.

- **Annexations or Jurisdictional Changes**: Administrative boundaries, such as city limits or county lines, can shift due to annexation or political decisions, affecting how census tracts are defined. For instance, if a city expands, its new boundaries may lead to new tracts or redrawing of existing tracts.

- **Geographic and Environmental Factors**: Natural events like flooding, changes in rivers, or other environmental factors can affect how boundaries are drawn. Similarly, tracts in regions with large geographic features (mountains, rivers) may remain stable compared to flat, highly developable land.

### Visualizing the Changes:
To further understand and explain these changes:
1. **Create Maps**: Visualizing the tracts on a map (using GIS software like ArcGIS) for both 2010 and 2020 can help show where boundaries have changed, where new tracts have been created, and where tracts have merged.
   
2. **Compare Population Data**: If available, compare population data for the 2010 and 2020 tracts. This can help identify whether the changes are due to shifts in population density (e.g., rapid growth in suburbs or depopulation in rural areas).

3. **Check Census Documentation**: The Census Bureau often provides documentation or reports on tract changes between censuses. These reports explain why certain tracts were created, removed, or modified, based on population changes and urban planning needs.

### Conclusion:
The changes in census tracts from 2010 to 2020 reflect dynamic population shifts and development patterns. The 38 tracts unique to 2010 likely represent areas that were merged or renamed, while the 74 new tracts in 2020 indicate areas of growth and development where the population has increased or been reallocated. Understanding these patterns can give insights into demographic and spatial changes over time.

```python
changes['popchangePercent'] = ((changes['POPTOTAL_2020'] - changes['POPTOTAL2010'])/changes['POPTOTAL2010'])*100
```

```python
# Summary statistics for the population percentage change
summary_stats = changes['popchangePercent'].describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt

# Histogram of population percentage change
plt.hist(changes['popchangePercent'], bins=20, edgecolor='black')
plt.title('Distribution of Population Change Percentages (2010-2020)')
plt.xlabel('Population Change Percent (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Identify tracts with large positive or negative changes
large_increases = changes[changes['popchangePercent'] > changes['popchangePercent'].mean() + 2 * changes['popchangePercent'].std()]
large_decreases = changes[changes['popchangePercent'] < changes['popchangePercent'].mean() - 2 * changes['popchangePercent'].std()]
```

```python
# Print or inspect the tracts with large population increases or decreases
print("Census tracts with large population increases:")
print(large_increases)
```

```python

print("Census tracts with large population decreases:")
print(large_decreases)
```

```python
import matplotlib.pyplot as plt

# Plotting population change as a choropleth map
changes.plot(column='popchangePercent', cmap='coolwarm', legend=True)
plt.title('Population Change Percent (2010-2020) by Census Tract')
plt.show()
```

```python
# Categorize population changes into groups based on the quartiles
def categorize_change(change):
    if change < 872.83:
        return 'Low Decline/Stable'
    elif 872.83 <= change < 1161.12:
        return 'Moderate Growth'
    elif 1161.12 <= change < 1492.88:
        return 'High Growth'
    else:
        return 'Very High Growth'


```

```python
# Apply the categorization function to the 'popchangePercent' column
changes['popchangeCategory'] = changes['popchangePercent'].apply(categorize_change)

# Analyze counts of each category
category_counts = changes['popchangeCategory'].value_counts()
print(category_counts)

# Optionally, visualize the counts
import matplotlib.pyplot as plt
category_counts.plot(kind='bar', color='skyblue')
plt.title('Census Tracts by Population Change Category')
plt.xlabel('Population Change Category')
plt.ylabel('Count of Census Tracts')
plt.show()
```
